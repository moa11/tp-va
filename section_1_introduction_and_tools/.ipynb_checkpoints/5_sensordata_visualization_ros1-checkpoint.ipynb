# Launch two terminals from the launcher

Launch two different terminals. In order to do that click on

**File >> New Launcher >> Terminal**

and execute the following commands in the terminals: 


## 1st Terminal

Open the first terminal tab in the browser and run the following code, 


```bash
source /opt/ros/noetic/setup.sh
```

```bash
cd acdc/section_1_introduction_and_tools
```

```bash
roslaunch websocket.launch
```

This commands will
- Setup the ROS environment
- Navigate to `/acdc/section_1_introduction_and_tools`
- Launch the launchfile `websocket.launch` to esatblish the connection between ROS and Zethus


## 2nd Terminal

Open the third terminal tab in the browser and run the following code, 


- Setup the ros environment
- Navigate to the /acdc/section_1_introduction_and_tools 
- Play the Rosbag `carla.bag` in the loop and start publishing the data as if data is being published from the sensors on the actual vehicle.

# Visualize sensor data with Zethus

Now, it is time to launch the Zethus from the launcher tab. Execute the following steps,

### 1. Click on - **File >> New Launcher >> Zethus**



### 2. Click on Add visualizations



- Select the topic you want to subscribe to
- Click on proceed, type the topic name
- Click **Add Visualization** it would start visualizing the data in Zethus



### 4. Final you should see following type of the visualization




### 5. Add Point Cloud visualization

- Select: Fixed Frame as `ego_vehicle/lidar`
- Click on **Add Visualization**
- Scroll down to **POINTCLOUD** and select `/carla/ego_vehicle/lidar` and then on proceed
- Try to increase the size of the point cloud to `0.5`

You should see the visualization of the recorded point cloud as follow



### 6. Checkout other messages types and try to visualize them with Zethus

The rosbag contains many different message types on different topics. You can also try to visualize the following topics

- `/carla/ego_vehicle/odometry`
- `/carla/ego_vehicle/waypoints`
